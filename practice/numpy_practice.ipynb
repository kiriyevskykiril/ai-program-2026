```python
import timeit
import numpy as np
```

```python
numRows = 300
numCols = 500
numIter = 1000
```

```python
timeit.timeit("sum(range(1000))")
```

```python
timeit.timeit("sum(range(1000))", number=1000)
```

```python
timeOnes = timeit.timeit(lambda: np.ones((numRows, numCols)), number=numIter)
print(timeOnes)
timeEmpty = timeit.timeit(lambda:np.empty((numRows, numCols)), number=numIter)
print(timeEmpty)

if timeOnes < timeEmpty:
    print(f'Generating array of ones is {timeEmpty / timeOnes} times faster!')
else:
    print(f'Generating empty array is {timeOnes / timeEmpty} times faster!')
```

```python
%timeit np.zeros(shape = (numRows, numCols))
%timeit np.full(shape=(numRows, numCols), fill_value=0.0)
```

```python
mA_shape = (1000, 1000)
mR_shape = (100, 100)

mA = np.random.randn(*mA_shape)
mR = np.zeros((100, 100))
```

```python
%%timeit
for ii, rows in enumerate(range(0, mA_shape[0], 10)):
    for jj, cols in enumerate(range(0, mA_shape[1], 10)):
        mR[ii, jj] = np.mean(mA[rows : rows + 10,cols : cols + 10])
```
