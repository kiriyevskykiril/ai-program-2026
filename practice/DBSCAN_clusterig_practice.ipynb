```python
# Import Packages

# General Tools
import numpy as np
import scipy as sp
import pandas as pd

# Machine Learning
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Miscellaneous
from platform import python_version
import random

# Typing
from typing import Callable, Dict, List, Optional, Self, Set, Tuple, Union
from numpy.typing import NDArray

# Visualization
import matplotlib.pyplot as plt

# Jupyter
from IPython import get_ipython
from ipywidgets import Dropdown, FloatSlider, IntSlider, Layout, SelectionSlider
from ipywidgets import interact
```

```python
# data generation

vNumSamples = [250, 250, 50]

mX0, _ = make_moons(vNumSamples[0], noise= 0.05)
mX1, _ = make_moons(vNumSamples[1], noise= 0.05)
mX1 = mX1 * [1, -1] + [0, 3]
mX2 = np.random.rand(vNumSamples[2], 2) * [4, 5] - [1.75, 2/3]
mX = np.r_[mX0, mX1, mX2]
vL = np.repeat(range(len(vNumSamples)), vNumSamples)

print(mX.shape)
```

```python
fig, ax = plt.subplots()

vNumSamples = mX.shape[0] #number of rows in mX matrix
vU = np.unique(vL) # unique values of label vector
numClusters = len(vU) # lenth of unique values vector, says number of clusters

for ii in range(numClusters):
    vIdx = vL == vU[ii] # indexes of cluster in the loop
    ax.scatter(mX[vIdx, 0], mX[vIdx, 1], label = ii)

ax.legend()
ax.grid()

plt.show()
```
