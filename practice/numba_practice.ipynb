```python
import random
```

```python
from numba import njit
```

```python

@njit
def MonteCarloPi(numSamples):
    num_in_circle = 0
    for i in range(numSamples):
        xVal = random.random()
        yVal = random.random()

        if ((xVal ** 2 + yVal **2) < 1.0):
            num_in_circle = num_in_circle + 1

    #print(f"Number of samples in circle is {num_in_circle}")

    return (4 * num_in_circle) / numSamples
    
```

```python
%%timeit
MonteCarloPi(1_000_000)
```

```python
def MonteCarloPi_Numpy(numSamples):
    rng = np.random.default_rng()
    vX = rng.random(numSamples)
    vY = rng.random(numSamples)

    np.sum((np.square(vX) + np.square(vY)) < 1.0)

    return (4 * np.sum((np.square(vX) + np.square(vY)) < 1.0)) / numSamples
```

```python
%%timeit
MonteCarloPi_Numpy(1_000_000)
```
