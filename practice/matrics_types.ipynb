```python
import numpy as np
```

```python
# PD positive definite 

A = np.array([[2, 0],
              [0, 3]])

# x.T @ A @ x > 0 for every x != 0

# Matrix is PD if all eigenvalues > 0

eigen_vals = np.linalg.eigvals(A)
print(eigen_vals)
```

```python
# PSD Positive Semi Definite 
# x.T @ A @ x >= 0

A = np.array([[1, 0],
             [0, 0]])

# for some x's x.T @ A @ x = 0
x = np.array([0, 1])
x.T @ A @ x
```
