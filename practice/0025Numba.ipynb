## Numba

[_Numba_](https://github.com/numba/numba) is a _JIT_ compiler for hot code based on _NumPy_ arrays.

This notebook shows the performance uplift in performance due to _Numba_.

<div class="alert alert-success">
Numba is very efficient when used properly!
</div>

<div class="alert alert-info">
Gains are made for hot code. For code which runs once, the overhead might cause no gains.
</div>

<div class="alert alert-danger">

## Example 001 - Sorting

Implement the sort function `SortArray()` (Ascending order) using one of the world simplest sorting algorithm:

> ![](https://i.imgur.com/jPpZoAW.png)
>
> See [The Simplest and Most Surprising Sorting Algorithm Ever](https://arxiv.org/abs/2110.01111).

The function will be implemented in 2 flavors:

1. Python + NumPy.
2. Numba + NumPy.

* <font color='red'>(**?**)</font> Does the algorithm sort the values in _ascending_ or _descending_ order?

````python
# Sorting: Python

#===========================Fill This===========================#
# 1. Implement the sorting algorithm in Python + NumPy.

def SortArrayPython( vA: NDArray ) -> NDArray:

    for i in range(len(vA)):
        for j in range(len(vA)):
            if (vA[j] > vA[i]):
                vA[j], vA[i] = vA[i], vA[j] #swapping
    
    return vA

#===============================================================#
````

````python
vA = np.random.randint(0, 8, size = (6))
print(vA)
SortArrayPython(vA)
````

````python
%%timeit
vA = np.random.randint(0, 8, size = (6))
SortArrayPython(vA)
````

````python
# Sorting: Numba

#===========================Fill This===========================#
# 1. Implement the sorting algorithm in Python + NumPy + Numba.

@njit
def SortArrayNumba( vA: NDArray ) -> NDArray:

    for i in range(len(vA)):
        for j in range(len(vA)):
            if (vA[j] > vA[i]):
                vA[j], vA[i] = vA[i], vA[j] #swapping
    
    return vA

#===============================================================#
````

````python
# Verify Implementation
numSamples = 1_000

vA = np.random.rand(numSamples)
vR = np.sort(vA) #<! Reference
````

````python
# Verify the Python Version
vB = vA.copy()
vB = SortArrayPython(vB)

print(f'The Python sorting version is verified: {np.all(vB == vR)}')
````

````python
# Verify the Numba Version
vB = vA.copy()
vB = SortArrayNumba(vB)

print(f'The Numba sorting version is verified: {np.all(vB == vR)}')
````

## Example 003 - Fused Loop

This section implements the function:

$$
f \left( x \right) = {\left( \log \left( \sqrt{\left| x \right| + 10} * 2 \right) + 2.5 \right)}^{2}
$$

The function is implemented on an array of numbers in 2 manners:

1. NumPy array operations.
2. A loop using Numba JIT.

````python
# Generate the Data

numRows = 2_000
numCols = 2_000

mX = np.random.rand(numRows, numCols)
````

````python
# NumPy Style Implementation

#===========================Fill This===========================#
# 1. Implement the function using NumPy.

def MyFunNumpy( mX: NDArray ) -> NDArray:
  
  mY = np.square(np.log(2 * np.sqrt(np.abs(mX) + 10)) + 2.5)
  
  return mY
#===============================================================#
````

````python
# Numba Style Implementation

#===========================Fill This===========================#
# 1. Implement the function using Numba.

@njit(fastmath = True)
def MyFunNumba( mX: NDArray ) -> NDArray:
  
  mY = np.square(np.log(2 * np.sqrt(np.abs(mX) + 10)) + 2.5)
  
  return mY
#===============================================================#
````

````python
# Run Time Comparison

%timeit MyFunNumpy(mX)
%timeit MyFunNumba(mX)
````
