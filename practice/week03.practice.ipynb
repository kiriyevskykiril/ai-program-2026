```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def single_exp():
    return 1 if random.random() < 2/3 else 0

def success_num(num_of_singe_exp=20):
    return sum([single_exp() for i in range(num_of_singe_exp)])

def pmf(n_times = 1000):
    results = [success_num(20) for i in range(n_times)]
    return results
```

```python
results = pmf(1000)
plt.hist(results, bins=12)
plt.show()
```

```python
# creation of dictionary
counts = {} #empty dictionary

for i in results:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1

results_arr = np.array(list(counts.items()))
results_arr = results_arr[results_arr[:,0].argsort()] #sortin
```

```python
results_arr = results_arr.astype('float')
results_arr.dtype
results_arr[:, 1] = (results_arr[:, 1] / 1000)
results_arr
```

```python
plt.stem(results_arr[:, 0], results_arr[:, 1])
plt.show()
```

```python
# Liklihood for k =10, N = 20
import math

p = np.arange(0, 1, 0.02)
# Liklehood to get k=10, from N = 20, with certain value of p 
L = math.comb(20, 10) * p **10 * (1-p)**10
```

```python
plt.plot(p, L)
plt.title("Likelihood")
plt.xlabel('p')
plt.show()
```
