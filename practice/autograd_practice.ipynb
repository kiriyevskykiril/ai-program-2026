```python
import autograd.numpy as anp
from autograd import grad
import matplotlib.pyplot as plt
from autograd import elementwise_grad as egrad
```

```python
def tanh(x):
    return (1.0 - anp.exp(-2 * x)) / (1.0 + anp.exp(-2 * x))
```

```python
x = anp.linspace(-anp.pi, anp.pi, 100)
y = tanh(x)
grad_tanh = grad(tanh)
#y_tag = [grad_tanh(i) for i in x]
y_tag = egrad(tanh)(x) 
```

```python
plt.plot(x, y)
plt.plot(x, y_tag)
plt.show()
```
