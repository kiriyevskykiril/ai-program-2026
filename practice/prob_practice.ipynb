```python
import random
```

```python
# function to simulate the coint toss
def flip_coin():
    return random.choice(['H', 'T'])

def simulate_coint_tosses(n):
    results = [flip_coin() for _ in range(n)] 
    return results

def probability_of_heads(results):
    heads_count = results.count('H')
    return heads_count / len(results)

def probability_of_tails(results):
    return 1 - probability_of_heads(results)
```

```python
n_tosses = 1000
results = simulate_coint_tosses(n_tosses)
prob_heads = probability_of_heads(results)
print(f"Out of {n_tosses} the probability of heads is aproximately {prob_heads}")
```

```python
prob_tails = probability_of_tails(results)
print(f"Out of {n_tosses} the probability of tails is aproximately {prob_tails}")
```

```python
def roll_die():
    return random.randint(1, 6)

def simulate_die_rolls(n):
    results = [roll_die() for _ in range(n)]
    return results

def probability_of_odd(results):
    odd_count = len([x for x in results if x % 2 != 0])
    return odd_count / len(results)

def prob_of_odd_or_greater_than_4(results):
    odd_count = len([x for x in results if x % 2 != 0])
    greater_than_4_count = len([x for x in results if x > 4])
    both_count = len([x for x in results if (x % 2 != 0) and (x > 4)])
    return (odd_count + greater_than_4_count - both_count) / len(results)

def roll_two_dice(): 
    return (roll_die(), roll_die())

def simulate_two_dice_rolls(n):
    return [roll_two_dice() for _ in range(n)]

def probability_two_fours(results):
    both_fours_counts = len([x for x in results if x == (4, 4) ])
    return both_fours_counts / len(results)

```

```python
n_rolls = 1000
die_results = simulate_die_rolls(n_rolls)
prob_odd = probability_of_odd(die_results)
print(f'Out of {n_rolls} rolls, the probability of rolling an odd number is {prob_odd}')
print(f"probability of rolling an odd number or greater than 4 {prob_of_odd_or_greater_than_4(die_results)}")
```

```python
n_two_rolls = 1000
two_dies_results = simulate_two_dice_rolls(n_two_rolls)
prob_two_fours = probability_two_fours(two_dies_results)
prob_two_fours = probability_two_fours(two_dies_results)
print(prob_two_fours)
```

```python
# Geometric destribution
def n_trials_to_get_four():
    result = roll_die()
    count = 1
    while result != 4:
        result = roll_die()
        count += 1
    return count
```

```python
trials = 10
results = [n_trials_to_get_four() for _ in range(trials)]
```

```python
import matplotlib.pyplot as plt

plt.scatter(range(trials), results)
plt.show()
```
