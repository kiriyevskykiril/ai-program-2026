```python
import pandas as pd
```

```python
mX13 = pd.read_csv('data/13features_matrix.csv', index_col=False)
mX13
```

```python
waka_dragon_df = pd.read_csv("data/waka_dragon_merged.csv")
vY_Imax = waka_dragon_df['Imax'].astype(float)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

X_scaled = pipeline.fit_transform(mX13)

dbscan = DBSCAN(
    eps=2.0,
    min_samples=5
)

clusters = dbscan.fit_predict(X_scaled)
```

```python
pd.DataFrame({
    'cluster': clusters,
    'Imax': vY_Imax
}).groupby('cluster')['Imax'].agg(['count','mean','std'])
```
