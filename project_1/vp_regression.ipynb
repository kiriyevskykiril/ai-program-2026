```python
import pandas as pd
```

```python
vp_df = pd.read_csv('data/output_vp_full_v4.csv')
vp_df.head(10)
```

```python
vp_df = vp_df[['CID', 'vp_mmhg_pubchem_clean']]
vp_df.dropna(subset=['vp_mmhg_pubchem_clean'], inplace=True)
print(vp_df.shape)
vp_df.rename(columns={'vp_mmhg_pubchem_clean' : 'vp_exp'}, inplace=True)
vp_df.head()
```

```python
# now let's how many experimental values we have in Waka data set
waka_df = pd.read_csv("data/waka_with_CID.csv")
waka_df.head()
```

```python
waka_vp = waka_df.merge(
    vp_df,
    on="CID",
    how='left'    
)

print(
    waka_vp["vp_exp"]
    .notna()
    .sum()
)
```

```python
dragon_df = pd.read_csv('data/raw_dragon_matrix.csv')
dragon_df.head()
```

```python
# Now, let's proceed with regression  models  attempts to predict the vapor pressure

# merging with dragon df

print(f"Initial shape of vp_df {vp_df.shape}")

vp_dragon_df = pd.merge(
    vp_df,
    dragon_df,
    how='left',
    on='CID',
)

print(f"Initial shape after merging: {vp_dragon_df.shape}")
print(f'Nimber of rows withou Dragon values: {vp_dragon_df.isna().sum()}')

# removing of rows without values
vp_dragon_df = vp_dragon_df.dropna()
print(f'After droping of rows without Dragon values wh have a shape of {vp_dragon_df.shape}')

# I need to delete rows where vp_exp <= 0
vp_dragon_df = vp_dragon_df.loc[vp_dragon_df['vp_exp'] > 0.0]
vp_dragon_df.head()
```

```python
vp_dragon_df['vp_exp'].min()
```

```python
mX_df_stage0 = vp_dragon_df.loc[:, 'MW':]
vY_ds = vp_dragon_df['vp_exp'].copy()
```

```python
# interesting to see how the vapor pressure is destributed
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(ncols=2, nrows=1, figsize = (12, 6))
ax[0].hist(vY_ds, bins= 30)
ax[0].set_title('Destribution of Vapor pressure values')
ax[0].set_xlabel('Vapor Pressure values in mm Hg')
ax[0].set_ylabel('Sample counts')
ax[1].hist(np.log10(vY_ds), bins= 30)
ax[1].set_title('Destribution of log10(VP)')
ax[1].set_xlabel('Log10(VP)')
ax[1].set_ylabel('Sample counts')

plt.show()
```

```python
# let's proceed with cleaning of feture matrix

#first we remove features with zero variance
from sklearn.feature_selection import VarianceThreshold

vt = VarianceThreshold(threshold=0) #object 

mX_stage1 = vt.fit_transform(mX_df_stage0) #returns NDArray

# mX1_df becomes the updated feature matrix
mX_df_stage1 = pd.DataFrame(
    mX_stage1,
    columns=mX_df_stage0.columns[vt.get_support()],
    index=mX_df_stage0.index
)

print(mX_df_stage1.shape)
```

```python
print(f'The number of reduced features: {(~vt.get_support()).sum()}')
print(f'The features which were removed: {mX_df_stage0.columns[~vt.get_support()]}')
```

```python
# now we want to reduce the features with high correaltion
# finfing of descriptors with correaltion higher than 0.95

# correlations matrix
corr_matrix = mX_df_stage1.corr().abs()

upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

to_drop = [
    column
    for column in upper.columns
    if any(upper[column] > 0.95) # May be the correlation threshold should be lower
]

print(f"Descriptors to remove: {len(to_drop)}")
```

```python
# Dropping of high correlated descriptors
mX_df_stage2 = mX_df_stage1.drop(columns=to_drop)
print(mX_df_stage2.shape)
```

```python
mX_df_stage2.shape
```

```python
vY = np.log10(vY_ds)
print(vY.shape)
```

```python
# Train XGBoost on All reamining descriptors

from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=1000,
    objective='reg:absoluteerror',
    max_depth=4,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

xgb.fit(mX_df_stage2, vY)
```

```python
importance = pd.Series(
    xgb.feature_importances_,
    index=mX_df_stage2.columns
).sort_values(ascending=False)
```

```python
importance.head(60)
```

```python

X60 = mX_df_stage2[importance.index[:60]]
```

```python
param_grid = {
    "n_estimators": [200, 500, 1000],
    "max_depth": [2, 3, 4],
    "learning_rate": [0.01, 0.03, 0.05],
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.5, 0.7, 0.9],
    "min_child_weight": [1, 3, 5],
    "reg_alpha": [0, 0.01, 0.1],
    "reg_lambda": [1, 3, 10]
}
```

```python
from sklearn.model_selection import KFold, cross_val_predict, GridSearchCV
```

```python
cv = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring="neg_mean_absolute_error",
    cv=cv,
    n_jobs=-1,
    verbose=2
)

grid.fit(X60, vY)

print("Best CV MAE:", -grid.best_score_)
print("Best params:")
print(grid.best_params_)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_absolute_error, r2_score

vY_pred_log = cross_val_predict(
    grid.best_estimator_,
    mX_df_stage2,
    vY,
    cv=cv,
    n_jobs=-1
)

vY_pred_mmhg = 10 ** vY_pred_log
vY_true_mmhg = 10 ** vY

mae_log = mean_absolute_error(vY, vY_pred_log)
r2_log = r2_score(vY, vY_pred_log)

mae_mmhg = mean_absolute_error(vY_true_mmhg, vY_pred_mmhg)
r2_mmhg = r2_score(vY_true_mmhg, vY_pred_mmhg)

print("CV MAE log10(VP):", mae_log)
print("R2 for log10(VP):", r2_log)

print("")
print("CV MAE VP mmHg:", mae_mmhg)
print("R2 for VP in mmHg :", r2_mmhg)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize = (12, 6))

ax[0].scatter(vY_true_mmhg, vY_pred_mmhg, alpha=0.5, label = 'VP in mmHg Estimation')
ax[0].plot(vY_true_mmhg, vY_true_mmhg, alpha=0.5, lw=1, linestyle = '--', color='k', label = 'Ground True')
ax[0].set_title('Calculated VP in mmHg from predicted log10(VP)')
ax[0].set_xlabel("True VP in mmHg")
ax[0].set_ylabel("Predicted vP in mmHg")
ax[0].legend()


ax[1].scatter(vY, vY_pred_log, alpha=0.5, label = 'Log10(VP) Estimation')
ax[1].plot(vY, vY, alpha=0.5, lw=1, linestyle = '--', color='k', label='Ground True')
ax[1].set_title('CV XGB prediction of Log10(VP)')
ax[1].set_xlabel("True Log10(VP)")
ax[1].set_ylabel("Predicted Log10(VP)")

ax[1].text(-7.0, 1.85, f"MAE: {mae_log:.2f}")
ax[1].text(-7.0, 1.45, f"R2: {r2_log:.2f}")
ax[1].legend()


plt.show()
```
