```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
waka_df = pd.read_csv('data/waka_dragon_merged.csv')
waka_df.head()
```

```python
fig, ax = plt.subplots(figsize = (16, 8))


ax.scatter(waka_df.loc[waka_df['intensity_class'] == 0, 'MW'], waka_df.loc[waka_df['intensity_class'] == 0, 'Imax'])
ax.scatter(waka_df.loc[waka_df['intensity_class'] == 1, 'MW'], waka_df.loc[waka_df['intensity_class'] == 1, 'Imax'])
ax.set_xlabel('MW')
ax.set_ylabel('Imax')
ax.set_title('MW vs Imax')
plt.show()
```
